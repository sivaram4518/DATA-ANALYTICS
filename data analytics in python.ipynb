```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.DataFrame({
    'marks':[78,56,78,65,45,46,45],
    'hours':[5,4,5,6,5,3,6]
})
print(data)
print("mean:",np.mean(data['marks']))
print("median:",np.median(data['marks']))
print("standard deviation:",np.std(data['marks']))
print("correlation:",np.corrcoef(data['marks'],data['hours']))

plt.scatter(data['hours'], data['marks'])
plt.title('Marks vs. Hours Studied')
plt.xlabel('Hours Studied')
plt.ylabel('Marks')
plt.show()
print()
plt.bar(data['hours'], data['marks'])
plt.title('Marks vs. Hours Studied')
plt.xlabel('Hours Studied')
plt.ylabel('Marks')
plt.show()
print()
mark_counts = data['marks'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(mark_counts, labels=mark_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Marks')
plt.axis('equal')
plt.show()

print()
plt.hist(data['marks'], bins=5, color='purple', edgecolor='black')
plt.title('Histogram of Marks')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show()
print()
plt.boxplot(data['marks'])
plt.title('Box Plot of Marks')
plt.ylabel('Marks')
plt.show()
print()
print("\nStatistical Summary:")
print(data.describe())
```
