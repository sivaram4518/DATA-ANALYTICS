```python
#1ST
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.DataFrame({
    'ordertime':[10,9,5,3,6,2],#hrs
    'deliverytime':[35,46,28,34,31,49],#mins
    'deliverypartner':['A','B','C','B','A','B'],
    'customerratings':[4,5,3,2,4,5],
    'area':['mpl','angallu','mpl','viswam','mits','mits'],
    'ordercount':[3,4,2,1,2,5]
})
print(df)
averagedeliverytime=df['deliverytime'].mean()
print('averagedeliverytime:',averagedeliverytime,'mins')
peakorderhours=[]
ordercount=df['ordercount']
ordertime=df['ordertime']
for i in ordercount.index:
 if ordercount[i] > 3:
  peakorderhours.append(ordertime[i])
print(f"peakorderhours at {peakorderhours}hrs")
mostdelayedareas = []
deliverytime = df['deliverytime']
area = df['area']
for i in deliverytime.index:
 if deliverytime[i] > 40:
  mostdelayedareas.append(area[i])
print('mostdelayedareas:',mostdelayedareas)
customerrating=df['customerratings']
topratedpartner=[]
for i in customerrating.index:
 if customerrating[i] > 4:
  topratedpartner.append(df['deliverypartner'][i])
print('topratedpartner:',topratedpartner)
for i in ordercount.index:
 if ordercount[i] > 4:
  print("assign more deliverypartners")
 elif deliverytime[i]>40 :
  print("assign more deliverypartners")
 else:
  print("assign less deliverypartners")
print()
plt.scatter(df['ordertime'],df['deliverytime'])
plt.xlabel('ordertime')
plt.ylabel('deliverytime')
plt.show()
print()
plt.bar(df['deliverypartner'],df['customerratings'])
plt.xlabel('deliverypartner')
plt.ylabel('customerrating')
plt.show()
print()
plt.pie(df['ordercount'],labels=df['area'])
plt.title('ordercount')
plt.show()

```

```python
#2ND
import pandas as pd
import statistics as st
import numpy as np
import matplotlib.pyplot as plt
df=pd.DataFrame({
'productpurchaged':['phone','book','slippers','food'],
'cartabandonment':['clothes','toys','pouch','oil'],
'customerratings':[4,5,3,4],
'paymentmethods':['upi','cod','creditcard','cod'],
'sales':[18,34,56,55],
 'revenue':[98000,34000,56000,55000],
 'customer':['a','b','c','d'],
  'customerrevenue':[10000,20000,30000,40000],
    'price':[20000,2000,800,650]
})
print(df)
#most sold
mostsoldproduct=df['productpurchaged'][df['sales'].idxmax()]
print('mostsoldproduct:',mostsoldproduct)

#payment method
mostpreferredpaymentmethod=st.mode(df['paymentmethods'])
print('mostpreferredpaymentmethod:',mostpreferredpaymentmethod)

#high revenue customer
highrevenuecustomer=df['customer'][df['customerrevenue'].idxmax()]
print('highrevenuecustomer:',highrevenuecustomer)

#poor rating
productwithpoorrating=df['productpurchaged'][df['customerratings'].idxmin()]
print('productwithpoorrating:',productwithpoorrating)
print()

plt.bar(df['productpurchaged'],df['sales'])
plt.xlabel('productpurchaged')
plt.ylabel('sales')
plt.title('product sales')
plt.show()
print()
plt.pie(df['customerratings'],labels=df['customer'])
plt.title('customer ratings')
plt.show()

```

```python
#3RD
import pandas as pd
import statistics as st
import numpy as np
import matplotlib.pyplot as plt
df=pd.DataFrame({
    'transactionamount':[1000,300.99,565.53,9000,5335.55,9486.46],
    'location':['mp','ap','up','hp','tn','goa'],
    'time':[10, 1, 3, 22,5, 3],
    'accountactivity':[10,13,20,9,14,22],
    'fraud':['no','yes','yes','no','yes','yes']
})
print(df)
unusual = df[(df['transactionamount'] % 1 !=0)]
print("\nUnusual Transactions:")
print(unusual)

high_risk = df[(df['transactionamount'] % 1 !=0) & (df['accountactivity'] >10)]
print("\nHigh Risk Accounts:")
print(high_risk)

fraud_cases = df[df['fraud'] == 'yes']

print("\nFraud Pattern Analysis")
print("Average Fraud Amount:", fraud_cases['transactionamount'].mean())
print("Average Fraud Activity:", fraud_cases['accountactivity'].mean())
print("\nFraud Time Distribution:")
print(fraud_cases['time'])


timedistribution = fraud_cases['time'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(timedistribution, labels=timedistribution.index, autopct='%1.1f%%')
plt.title('Fraud Cases Time Distribution')
plt.show()

print()
plt.scatter(df['time'],df['transactionamount'])
plt.xlabel('time')
plt.ylabel('transactionamount')
plt.title('transactionamount')
plt.show()
print()

plt.bar(df['location'],df['transactionamount'])
plt.xlabel('location')
plt.ylabel('transactionamount')
plt.title('transactionamount')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    'Product':['P1','P2','P3','P4','P5'],
    'DailySales':[120,15,80,10,60],
    'ProductStock':[50,200,40,150,70],
    'SeasonalDemand':['High','Low','High','Low','Medium']
}

df = pd.DataFrame(data)
print(df)

category = []

for i in range(len(df)):
    if (df['DailySales'][i] > 50 and df['ProductStock'][i] < 100 and(df['SeasonalDemand'][i] == 'High' or df['SeasonalDemand'][i] == 'Medium')):
        category.append("Fast Moving")

    elif (df['DailySales'][i] <= 50 and df['ProductStock'][i] >= 100 and df['SeasonalDemand'][i] == 'Low'):
        category.append("Slow Moving")

    else:
        category.append("Moderate Moving")

df['Category'] = category

print(df)

print("\nFast Moving Products:")
for i in range(len(df)):
    if df['Category'][i] == "Fast Moving":
        print(df['Product'][i])

print("\nSlow Moving Products:")
for i in range(len(df)):
    if df['Category'][i] == "Slow Moving":
        print(df['Product'][i])

print("\nOut of stock Products:")
for i in range(len(df)):
  if df['ProductStock'][i]==0:
    print(f"{df['Product'][i]} is out of stock")
print("no product is out of stock")
print()

plt.bar(df['Product'],df['DailySales'])
plt.xlabel('Product')
plt.ylabel('DailySales')
plt.title('DailySales')
plt.show()
plt.pie(df['ProductStock'],labels=category)
plt.title('ProductStock')
plt.show()
```
