PRESCRIPTIVE ANALYSIS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.DataFrame({
    'patientname':["p1","p2","p3","p4","p5"],
    'age':[23,45,43,67,56],
    'condition':['normal','lowserious','critical','critical','normal'],
    'oxygenlevel':[87,67,54,45,84],

    })
print(df)
patientname=df['patientname']
availablebeds=3
print("availablebeds",availablebeds)
patients=df[ (df['condition']=='critical') | (df['oxygenlevel']<50)]
print(patients)
for i in patients.index:
  patient_row=patients.loc[i]
  if availablebeds>0:
    print(f"bed allocated for {patient_row['patientname']}")
    availablebeds-=1
  else:
    print(f"bed not allocated for {patient_row['patientname']}")
print()
plt.bar(patientname,df['age'])
plt.xlabel('patientname')
plt.ylabel('age')
plt.title('age of patients')
plt.show()
print()
plt.bar(patientname,df['oxygenlevel'])
plt.xlabel('patientname')
plt.ylabel('oxygenlevel')
plt.title('oxygenlevel of patients')
plt.show()
print()
plt.pie(df['oxygenlevel'],labels=df['condition'])
plt.title('condition of patients')
plt.show()
print()
```
