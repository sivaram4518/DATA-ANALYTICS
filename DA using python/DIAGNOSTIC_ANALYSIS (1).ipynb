```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.DataFrame({
    'productname':['iphone','samsung','realme','oppo','vivo'],
    'sales':[122,345,432,213,434],
    'price':[60000,50000,40000,30000,20000]
})
print(df)
df_sorted_sales = df.sort_values(by='sales', ascending=False)
df_sorted_price = df.sort_values(by='price', ascending=False)
plt.figure(figsize=(10, 6))
plt.plot('productname', 'sales', data=df_sorted_sales)
plt.title('Sales by Product')
plt.xlabel('Product Name')
plt.ylabel('Sales')
plt.show()
print()
plt.figure(figsize=(10, 6))
plt.bar('productname', 'price', data=df_sorted_price)
plt.title('Price by Product')
plt.xlabel('Product Name')
plt.ylabel('Price')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.DataFrame({
    'custname':['siva','shanmukh','sunil','vardhan','nandha'],
    'internetspeed':['low','high','high','high','low'],
    'paymentfailure':['yes','no','no','no','yes'],
    'deliverycharges':[32,49,67,87,86],
    'purchasecompleted':['no','yes','yes','no','no']
})
print(df)
print()
total_abandnment=df['purchasecompleted'].value_counts()['no']
print("total_abandnment:",total_abandnment)

internet_diagnosis=df['internetspeed'].value_counts()['low']
print("internet_diagnosis:",internet_diagnosis)

payment_diagnosis=df['paymentfailure'].value_counts()['yes']
print("payment_diagnosis:",payment_diagnosis)

delivery_diagnosis=(df['deliverycharges'] > 50).sum()
print("delivery_diagnosis:",delivery_diagnosis)
print()
print("ROOT CAUSES ARE:")
for i in range(len(df['custname'])):
  if df['internetspeed'][i]=='low':
    print("low internet speed")
  elif df['paymentfailure'][i]=='yes':
    print("payment failed")
  elif df['deliverycharges'][i]>70:
    print("high delivery charges")


plt.bar('custname','deliverycharges',data=df)
plt.title('deliverycharges by custname')
plt.xlabel('custname')
plt.ylabel('deliverycharges')
plt.show()
print()
plt.pie(df['deliverycharges'], labels=df['custname'] )
plt.title('deliverycharges by custname')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('sales_data_sample (1).csv', encoding='latin1')
print(df)
print()
total_abandnment=df['STATUS'].value_counts()['Cancelled']
print("total_abandnment:",total_abandnment)

internet_diagnosis=(df['internetspeed']>200).sum()
print("internet_diagnosis:",internet_diagnosis)

payment_diagnosis=(df['paymentfailure']>2).sum()
print("payment_diagnosis:",payment_diagnosis)

delivery_diagnosis=(df['deliverycharges'] > 4000).sum()
print("delivery_diagnosis:",delivery_diagnosis)
print()
internet_count = 0
payment_count = 0
delivery_count = 0

for i in range(len(df['CUSTOMERNAME'])):

    if df['internetspeed'][i] > 200:
        internet_count += 1

    elif df['paymentfailure'][i] > 3:
        payment_count += 1

    elif df['deliverycharges'][i] > 4000:
        delivery_count += 1

print("ROOT CAUSES ARE:")
print("low internet speed:", internet_count)
print("payment failed:", payment_count)
print("high delivery charges:", delivery_count)
print()
plt.figure(figsize=(30, 12))
plt.bar(df['PRODUCTLINE'], df['PRICEEACH'])
plt.title('PRICE OF PRODUCT')
plt.xlabel('PRODUCT LINE')
plt.ylabel('PRICE EACH')
plt.show()
print()
ROOT_CAUSES = [internet_count, payment_count, delivery_count]
plt.pie(ROOT_CAUSES, labels=['low internet speed', 'payment failed', 'high delivery charges'])
plt.title('ROOT CAUSES')
plt.show()
print()



```
