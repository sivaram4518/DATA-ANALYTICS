```python
# Smart Fitness Center Descriptive Analysis

import statistics
import matplotlib.pyplot as plt
member_names=["siva","ram","sunil","vardhan","shanmukh","nandha"]
age=[20,22,21,19,20,23]
workout_hours=[1,2,1.5,2.5,2,4]
calories_burned=[500,600,550,600,400,500]
print("------fitness center dataset------")
for i in range(len(member_names)):
    print(member_names[i],
             age[i],
    workout_hours[i],
    calories_burned[i])
average_workout=sum(workout_hours)/len(workout_hours)
print("average workout hours:",average_workout)
max_calories=max(calories_burned)
print("maximum calories burned:",max_calories)
min_calories=min(calories_burned)
print("minimum calories burned:",min_calories)
median_calories=statistics.median(calories_burned)
print("median calories burned:",median_calories)
std_dev=statistics.stdev(calories_burned)
print("standard deviation:",std_dev)
index=calories_burned.index(max_calories)
print("most active member=",member_names[index])
plt.bar(member_names,calories_burned)
plt.xlabel("member names")
plt.ylabel("calories burned")
plt.title("member names vs calories burned")
plt.show()
print()
plt.pie(calories_burned,labels=member_names)
plt.title("calories distribution")
plt.show()
print()
plt.scatter(age,calories_burned)
plt.xlabel("age")
plt.ylabel("calories burned")
plt.title("age vs calories burned")
plt.show()
print()
plt.hist(workout_hours)
plt.xlabel("workout_hours")
plt.ylabel("frequency")
plt.title("workout_hours distribution")
plt.show()
print()



```
