```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.DataFrame({
    'waterconsumption':[10,13,14,15,16,18,17],#litres
    'days':['mon','tue','wed','thu','fri','sat','sun']
})
print(df)
waterconsumption=df['waterconsumption']
averagewaterconsumption=df['waterconsumption'].mean()
print("averagewaterconsumption:",averagewaterconsumption)
total_water_usage = df['waterconsumption'].sum()
print("Total water usage over the week:", total_water_usage, "litres")
df['daily_increase'] = df['waterconsumption'].diff()
avg = df['daily_increase'].mean()
print("Average daily increase in water usage:", avg)
increased_usage_days = df[df['daily_increase'] > 0]
print("Days with increased water usage:")
print(increased_usage_days)
sum_of_increased_usage = increased_usage_days['daily_increase'].sum()
print("Sum of daily increased water usage:",sum_of_increased_usage)
tankcapacity=15

shortage_occurred = False
for consumption in waterconsumption:
  if consumption > tankcapacity:
    shortage_occurred = True
    break
if shortage_occurred:
    print("Shortage occurred during the week.")
else:
    print("No shortage occurred during the week.")

print()
plt.plot(df['days'], df['waterconsumption'], marker='o')
plt.xlabel("Days")
plt.ylabel("Water Consumption (litres)")
plt.title("Weekly Water Consumption")
plt.grid(True)
plt.show()
print()
plt.bar(df['days'], df['waterconsumption'])
plt.xlabel("Days")
plt.ylabel("Water Consumption (litres)")
plt.title("Weekly Water Consumption")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('weather.csv')
temperature = df['MinTemp']
average_temperature = df['MinTemp'].mean()
print("Average Temperature:", average_temperature)
total_temperature = df['MinTemp'].sum()
print("Total Temperature:", total_temperature)
df['daily_difference'] = df['MinTemp'].diff()
avg_difference = df['daily_difference'].mean()
print("Average Daily Difference:", avg_difference)
increased_temp_days = df[df['daily_difference'] > 0]
print("Days with increased temperature:")
print(increased_temp_days)
sum_increase = increased_temp_days['daily_difference'].sum()
print("Sum of increased temperature:", sum_increase)
threshold = 35
high_temp_occurred = False
for temp in temperature:
    if temp > threshold:
        high_temp_occurred = True
        break
if high_temp_occurred:
    print("High temperature occurred.")
else:
    print("No high temperature occurred.")
print()

plt.plot(df.index, df['MinTemp'], marker='o')
plt.xlabel("Days")
plt.ylabel("Minimum Temperature")
plt.title("Daily Minimum Temperature")
plt.grid(True)
plt.show()

print()

plt.bar(df.index, df['MinTemp'])
plt.xlabel("Days")
plt.ylabel("Minimum Temperature")
plt.title("Daily Minimum Temperature")
plt.show()
```
