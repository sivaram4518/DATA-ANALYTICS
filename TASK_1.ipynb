```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=pd.DataFrame({
    'marks':[78,56,78,65,45,46,45],
    'hours':[5,4,5,6,5,3,6]
})

print(data)

print("Mean:",np.mean(data['marks']))
print("Median:",np.median(data['marks']))
print("Standard Deviation:",np.std(data['marks']))
print("Correlation:",np.corrcoef(data['marks'],data['hours']))

# Scatter Plot
plt.scatter(data['hours'],data['marks'])
plt.title("Hours vs Marks")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.show()

# Bar Chart
plt.bar(data['hours'],data['marks'])
plt.title("Hours vs Marks")
plt.show()

# Histogram
plt.hist(data['marks'],bins=5)
plt.title("Marks Distribution")
plt.show()

# Box Plot
plt.boxplot(data['marks'])
plt.title("Marks Boxplot")
plt.show()

print(data.describe())
```
